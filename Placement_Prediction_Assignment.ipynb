#Import The Libraries

```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

#Import The Dataset

```python
dataset = pd.read_csv('/content/complete_final_placement_data_cleaned.csv')
```

```python
dataset.columns.to_list()
```

```python
x = dataset.iloc[:, 0:11].values
```

```python
y = dataset.iloc[:, 11].values
```

#Feature Scaling

```python
from  sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
X = sc.fit_transform(x)
```

#Splitting the dataset into the training and test sets

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

#Training The Logistic Regression Model on the Training Set

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(X_train, y_train)
```

#Predicting The Test Set Results For Logisitic Regression

```python
y_pred = classifier.predict(X_test)
classifier.score(X_train, y_train)
```

#Making The Confusion Matrix And Finding Accuracy For Logistic Regression

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
sum(np.diag(cm))
acc = (sum(np.diag(cm))/len(y_test))
acc
```

#Plot The Placement Status For Train And Test Data Using Bar Plot For Logistic Regression Model Which Gives The Highest Accuracy.

```python
# Bar plot for Placement Status
placement_counts = dataset.iloc[:, 11].value_counts()  # Count placement status
plt.figure(figsize=(8, 5))
plt.bar(['Not Placed (0)', 'Placed (1)'], placement_counts, color=['red', 'blue'])
plt.title('Placement Status Bar Plot')
plt.xlabel('Placement Status')
plt.ylabel('Count')
plt.show()

```
